### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}
```

```python
# Display the first document in the results using pprint
first_hygiene_20 = establishments.find_one(query)
pprint(first_hygiene_20)
```

```python
# Convert the result to a Pandas DataFrame
import pandas as pd
results = establishments.find(query)
df_hygiene_20 = pd.DataFrame(list(results))
```

```python
# Display the number of rows in the DataFrame
No_rows= len(df_hygiene_20)

print("Number of rows in the DataFrame for hygiene score of 20:", No_rows)

```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "LocalAuthorityName": "London",
    "scores.RatingValue": {"$gte": 4}
}

# Use count_documents to display the number of documents in the result
count_london_rating_4 = establishments.count_documents(query)
print("Number of establishments in London with RatingValue >= 4:", count_london_rating_4)
```

```python
# Display the first document in the results using pprint
first_london_rating_4 = establishments.find_one(query)
pprint(first_london_rating_4)
```

```python
# Convert the result to a Pandas DataFrame
results = establishments.find(query)
df_london_rating_4 = pd.DataFrame(list(results))
```

```python
No_rows= len(df_london_rating_4)

print("Number of rows in the DataFrame for London establishments with RatingValue >= 4:", No_rows)
```

```python
# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame for London establishments with RatingValue >= 4:")
print(df_london_rating_4.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
# Convert latitude and longitude to floats
latitude = 51.49014200
longitude = 0.08384000

# Define the degree_search
degree_search = 0.01

# Create the query
query_top_5_rating_5 = {
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search},
    "scores.RatingValue": 5
}
```

```python
sort_top_5_rating_5 = [("scores.Hygiene", -1)]  # 1 for ascending order
```

```python
# Find the top 5 establishments matching the criteria and sorted by hygiene score
cursor_top_5_rating_5 = list(establishments.find(query_top_5_rating_5).sort(sort_top_5_rating_5).limit(5))
pprint(cursor_top_5_rating_5)
```

```python
# Convert result to Pandas DataFrame
df_top_5_rating_5 = pd.DataFrame(list(cursor_top_5_rating_5))
df_top_5_rating_5.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Print the first 10 results
pipeline_hygiene_0 = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Execute the aggregation pipeline and print the results
agg_result_hygiene_0 = list(establishments.aggregate(pipeline_hygiene_0))
print("Number of establishments with hygiene score of 0 in each Local Authority area:")
pprint(agg_result_hygiene_0)
```

```python
# Convert the result to a Pandas DataFrame
df_hygiene_0 = pd.DataFrame(agg_result_hygiene_0)
# Display the number of rows in the DataFrame
No_row = len(df_hygiene_0)
print("Number of rows in the DataFrame for establishments with hygiene score of 0 in each Local Authority area:", No_row )

# Display the first 10 rows of the DataFrame
df_hygiene_0.head(10)
```
